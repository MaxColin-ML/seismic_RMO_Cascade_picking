```python
import sys
sys.path.append('.')
sys.path.append('..')
sys.path.append('../..')
import numpy as np
import os
from config.data_path import data_dict
from utils.plot_func import wiggle_plot, heatmap_plot


def visual_func(group, id, save_root):
    save_path = os.path.join(save_root, group, id)
    gth_1 = np.load(os.path.join(data_dict[group], 'gth', 'gth_%s.npy'%id)).T.astype(np.float32)
    lab_1 = np.load(os.path.join(data_dict[group], 'lab', 'lab_%s.npy'%id), allow_pickle=True).item()
    os.makedirs(os.path.join(save_root, group), exist_ok=True)
    wiggle_plot(gth_1, lab_1, size=(4, 10), save_path='%s_ori.pdf' % save_path)
    heatmap_plot(gth_1, lab_1, size=(4, 10), save_path='%s_ori_hp.pdf' % save_path)
    wiggle_plot(gth_1, None, size=(4, 10), save_path='%s_ori_wo_lab.pdf' % save_path)
    heatmap_plot(gth_1, None, size=(4, 10), save_path='%s_ori_wo_lab_hp.pdf' % save_path)
```

```python
visual_dict = {
    'syn_us_train': ['Line0_Cdp0', 'Line0_Cdp193', 'Line0_Cdp6001'],
    'syn_us_valid': ['Line0_Cdp0', 'Line0_Cdp193', 'Line0_Cdp603'],
    'syn_us_test': ['Line0_Cdp0', 'Line0_Cdp193', 'Line0_Cdp603']
}
save_root = 'fig'
for group, id_list in visual_dict.items():
    for id in id_list:
        visual_func(group, id, save_root)
```
