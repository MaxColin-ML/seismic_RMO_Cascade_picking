```python
import os 
import numpy as np

def plot_predict(res_path, root_folder, sample_name):
    os.makedirs(root_folder, exist_ok=True)
    import matplotlib.pyplot as plt
    res_dict = np.load(res_path, allow_pickle=True).item()

    # curve final
    fig, ax = plt.subplots(figsize=(2, 20))
    ax.imshow(res_dict['gth_msfeat'][0], cmap='binary')
    for curve in res_dict['curve_auto'].values():
        ax.plot(curve[:, 0], curve[:, 1], c='red', lw=0.5)
    for spine in ax.spines.values():
        spine.set_visible(False)
    # ax.invert_yaxis()
    fig.savefig(os.path.join(root_folder, 'curve_final_%s.png' % sample_name), dpi=300, bbox_inches='tight', pad_inches=0.0001)
    plt.close()
```
